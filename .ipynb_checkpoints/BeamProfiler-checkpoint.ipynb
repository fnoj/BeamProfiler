```python
import sys
import os
from PyQt4.QtGui import *
from PyQt4.QtCore import *
import PyQt4.QtCore as QtCore
import PyQt4.QtGui as QtGui
from scipy import ndimage
from PIL import Image      
import numpy as np         
import matplotlib.pyplot as plt 
from matplotlib.pylab import *
from matplotlib.colors import LogNorm
import numpy.linalg as la
from mpl_toolkits.mplot3d import Axes3D
from scipy.optimize import curve_fit


class QGui(QtGui.QWidget):
	def __init__(self, parent=None):
		QtGui.QWidget.__init__(self,parent)
		
		self.vBox = QVBoxLayout()
		self.hBoxB = QHBoxLayout()
		self.hBoxP = QHBoxLayout()
		self.hBoxR = QHBoxLayout()
				
		self.button = QtGui.QPushButton('&Capture')
		self.button.clicked.connect(self.calculate)
		self.e1 = QLineEdit()
		self.e1.resize(100,25)
		self.e1.setFont(QFont("Arial",11))
		self.labelfname = QtGui.QLabel('Filename:')			
			
		self.hBoxB.addWidget(self.button)			
		self.hBoxB.addWidget(self.labelfname)
		self.hBoxB.addWidget(self.e1)
		self.hBoxB.addWidget(self.e1)
		self.hBoxB.addWidget(self.e1)

		self.profile = QtGui.QLabel()
		self.profile.setFixedWidth(270)
		self.profile.setFixedHeight(270)
		self.profile.setScaledContents(True)
		
		self.fit3d = QtGui.QLabel()
		self.fit3d.setFixedWidth(270)
		self.fit3d.setFixedHeight(270)
		self.fit3d.setScaledContents(True)
	
		self.hBoxP.addWidget(self.profile)			
		self.hBoxP.addWidget(self.fit3d)			
		
		self.profile2 = QtGui.QLabel()
		self.profile2.setFixedWidth(270)
		self.profile2.setFixedHeight(270)
		self.profile2.setScaledContents(True)
		
		self.fit3d2 = QtGui.QLabel()
		self.fit3d2.setFixedWidth(270)
		self.fit3d2.setFixedHeight(270)
		self.fit3d2.setScaledContents(True)
		
		self.hBoxR.addWidget(self.profile2)			
		self.hBoxR.addWidget(self.fit3d2)			
		
		self.vBox.addLayout(self.hBoxP)
		self.vBox.addLayout(self.hBoxR)
		self.vBox.addLayout(self.hBoxB)
		
		self.setLayout(self.vBox)
		
	
	def calculate(self):
		self.name = self.e1.text()
		#os.system(str('sh Capture.sh '+self.name))
		analice(self.name)
		self.profile.setPixmap(QtGui.QPixmap(str("./Capture/"+self.name+".jpg")))
		self.profile.update()
		self.fit3d.setPixmap(QtGui.QPixmap(str("./Fitting/3dfit.jpg")))
		self.fit3d.update()
		self.profile2.setPixmap(QtGui.QPixmap(str("./Fitting/xfit.jpg")))
		self.profile2.update()
		self.fit3d2.setPixmap(QtGui.QPixmap(str("./Fitting/yfit.jpg")))
		self.fit3d2.update()
		
			
	def get_fname(self):
		self.name=self.e1.text()
		return self.name

def analice(name):
	I = Image.open(str("./Capture/"+name+".jpg"))
	I1=I.convert('L') # convierte a escala de grises
	#I1.show()
	a=np.asarray(I1,dtype=np.uint8)/255.0 #convierte I1 en una matriz normalizada con /255
	N=a.shape[0]
	M=a.shape[1]
	max=np.max(a)
	x=np.arange(M)
	y=np.arange(N)
	X,Y = meshgrid(x,y)
   
	A=[]
	B=[]
	for i in range(N):
	    for j in range(M):
	        A.append([i,j,a[i][j]])
	        if a[i][j] ==max:
	              B.append([i,j,a[i][j]])
        
	#Image(plt)
	#print len(t)
	B = np.asarray(B)
	A = np.asarray(A)
	#sacar puntos max y min en x y y
	
	xm=int(abs(np.max (B[:,0])-np.min(B[:,0]))/2)+np.min(B[:,0])
	ym=int(abs(np.max (B[:,1])-np.min(B[:,1]))/2)+np.min(B[:,1])
	C=[]
	D=[]
	for i in range(N*M):
	    if A[i,0]==xm:
	        C.append([A[i,1],A[i,2]])
	    if A[i,1]==ym:
	        D.append([A[i,0],A[i,2]])            
	
	C=np.asarray(C)  
	D=np.asarray(D)  
	#fit de gaussiana
	
	def func(x, Af, Bf, Cf):
	    return Af * np.exp(-(x-Bf)**2/(2*Cf**2)) 
	
	(Af, Bf, Cf), _ = curve_fit(func, D[:,0], D[:,1])
	fig = figure(figsize=(12,10))
	ax1 = fig.gca()
	plt.plot(D[:,0],D[:,1],'.')
	plt.plot(D[:,0],func(D[:,0], Af, Bf, Cf))
	ax1.set_xlabel("x")
	ax1.set_ylabel("Intensidad")
	#print(Af, Bf, Cf)
	fig.savefig('./Fitting/xfit.jpg') 
	
	(Af, Bf, Cf), _ = curve_fit(func, C[:,0], C[:,1])
	fig = figure(figsize=(12,10))
	#ax1 = fig.add_subplot(222)
	plt.plot(C[:,0],C[:,1],'.')
	plt.plot(C[:,0],func(C[:,0], Af, Bf, Cf))
	ax1.set_xlabel("y")
	ax1.set_ylabel("Intensidad")
	#print(Af, Bf, Cf)
	fig.savefig('./Fitting/yfit.jpg') 
	
	fig=figure(figsize = (20,15))
	ax = fig.gca(projection='3d')
	
	figura = ax.plot_surface(X,Y,a,cmap = cm.jet,linewidth = 0)
	#ax.plot(B[:,0],B[:,1],B[:,2],'b-',alpha=0.5)
	fig.colorbar(figura, shrink=0.5, aspect=5)  
	ax.set_xlabel("x")
	ax.set_ylabel("y")
	ax.set_zlabel("Intensidad")
	ax.view_init(elev=45, azim=45)
	
	#cset = ax.contour(losX, losY, new/80, zdir='z', offset=-3, cmap=cm.jet)
	cset = ax.contour(X, Y, a, zdir='x', offset=-20, cmap=cm.jet)
	cset = ax.contour(X, Y, a, zdir='y', offset=-20, cmap=cm.jet)
	fig.savefig('./Fitting/3dfit.jpg') 
	#
	#imshow(a,cmap=cm.gist_rainbow_r)
	#colorbar()
	show()

				
def main():
	a = QApplication(sys.argv)    
	a.setWindowIcon(QtGui.QIcon('./Icon/BP_Icon.png'))
	w = QWidget()
	w.resize(560, 600)
	w.setWindowTitle("Beam Profiler") 
	analice("lena")
	#take_b = QGui(w)
	#take_b.show()
	
	#w.show()
	sys.exit(a.exec_())
    
if __name__=='__main__':
	main()
```
